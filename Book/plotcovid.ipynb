```python
covid_days = ["Day1","Day2","Day3","Day4","Day5"]
covid_cont = [83,128,130,133,153]
```

```python
covid_days.extend(("Day6","Day7","Day8","Day9","Day10"))
covid_cont.extend((189,205,231,261,275))
```

```python
from matplotlib import pyplot as plt 
plt.plot(covid_days,covid_cont,color="red", marker='o', linestyle='solid')
plt.title("Aumento De Contaminação ")
plt.show()
```

```python
xl = [i + 0.1 for i,_ in enumerate(covid_days)]
plt.bar(xl,covid_cont)
plt.title("Covid Days")
plt.xticks([i + 0.1 for i,_ in enumerate(covid_days)],covid_days)
```
